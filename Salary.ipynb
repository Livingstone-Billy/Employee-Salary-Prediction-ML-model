```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Salary_Data.csv")
data.head()
```

```python
#visualizing the data
plt.figure(figsize=(16,10))
sns.relplot(x="YearsExperience",y="Salary",kind='line',data=data)
```

```python
plt.figure(figsize=(25,10))
sns.barplot(x="YearsExperience",y="Salary",data=data)
```

```python
plt.scatter("YearsExperience","Salary")
```

```python
data.describe()
```

```python
#Building the model
from sklearn.model_selection import train_test_split
```

```python
X = data[["YearsExperience"]].values
y = data[["Salary"]]
y = np.array(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()
regr.fit(X_train,y_train)
predictions = regr.predict(X_test)
acc=regr.score(X_test,y_test)
```

```python
print("Prediction: ",predictions)
print("Accuracy: ",acc)
```

```python
import pickle
pickle.dump(regr,open('Salary.pkl','wb'))
```
